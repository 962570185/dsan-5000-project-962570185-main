```python
# import data
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns


file_name = '/Users/wan/Desktop/Assignment/dsan-5000-project-962570185-main/processed_file.csv'
df=pd.read_csv(file_name)  
print(df.shape)
```

```python
# Data Understanding
print(f'The set project objective: find the features that relevant to the movie rating.') 

# data types
print(f'The data types of the file:')
print(df.dtypes)
# Title               text
# Year                 numerical
# Summary             text
# Short Summary       text
# Genres              categorical
# IMDB ID             categorical
# Runtime              numerical
# YouTube Trailer     object
# Rating             numerical
# Movie Poster        object
# Director            object
# Writers             object
# Cast                object

# potential relationships
print(f'The potential factors related to the rating may be the Director and the Writers')
```

```python
# Descriptive Statistics

# For numerical variables
year = df['Year'].to_numpy()
year_mean = np.mean(year)
year_median = np.median(year)
year_mode = stats.mode(year).mode
year_std = np.std(year)
year_var = np.var(year)
print(f'The statistics for numerical variable: Year')
print(f'Mean: {year_mean}')
print(f'Median: {year_median}')
print(f'Mode: {year_mode}')
print(f'Standard deviation: {year_std}')
print(f'Variance: {year_var}')

run_time = df['Runtime'].to_numpy()
run_time_mean = np.mean(run_time)
run_time_median = np.median(run_time)
run_time_mode = stats.mode(run_time).mode
run_time_std = np.std(run_time)
run_time_var = np.var(run_time)
print(f'The statistics for numerical variable: Runtime')
print(f'Mean: {run_time_mean}')
print(f'Median: {run_time_median}')
print(f'Mode: {run_time_mode}')
print(f'Standard deviation: {run_time_std}')
print(f'Variance: {run_time_var}')

rating = df['Rating'].to_numpy()
rating_mean = np.mean(rating)
rating_median = np.median(rating)
rating_mode = stats.mode(rating).mode
rating_std = np.std(rating)
rating_var = np.var(rating)
print(f'The statistics for numerical variable: Rating')
print(f'Mean: {rating_mean}')
print(f'Median: {rating_median}')
print(f'Mode: {rating_mode}')
print(f'Standard deviation: {rating_std}')
print(f'Variance: {rating_var}')


# For categorical variables 
# frequency distributions
# bar charts to visualize data distribution.
# 
frequency_distribution = df['Genres'].value_counts()  #  'YourColumn' 

# 
print(frequency_distribution)

# 
frequency_distribution.plot(kind='bar')
plt.xlabel('Categories')  # 
plt.ylabel('Frequency')
plt.title('Frequency Distribution of YourColumn')  # 
plt.show()
```

```python
# Data Visualization

# Histogram of the Year
plt.figure(figsize=(10, 6))
plt.hist(df['Year'])  
plt.title('Histogram of Year')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of the Runtime
plt.figure(figsize=(10, 6))
plt.hist(df['Runtime'])  
plt.title('Histogram of Runtime')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()


# Histogram of the Rating
plt.figure(figsize=(10, 6))
plt.hist(df['Rating'])  
plt.title('Histogram of Rating')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Box plots of Genres by Year
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Genres', y='Year')  
plt.title('Box Plot of Genres by Year')
plt.show()

# Box plots of Genres by Runtime
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Genres', y='Runtime')  
plt.title('Box Plot of Genres by Runtime')
plt.show()

# Box plots of Genres by Rating
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Genres', y='Rating')  
plt.title('Box Plot of Genres by Rating')
plt.show()

# Scatter plots Year vs Runtime
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Year', y='Runtime')  
plt.title('Scatter Plot of Year vs Runtime')
plt.show()

# Scatter plots Year vs Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Year', y='Rating')  
plt.title('Scatter Plot of Year vs Rating')
plt.show()

# Scatter plots Runtime vs Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Runtime', y='Rating')  
plt.title('Scatter Plot of Runtime vs Rating')
plt.show()
```

```python
# Correlation Analysis

# heatmaps 
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()  # 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
